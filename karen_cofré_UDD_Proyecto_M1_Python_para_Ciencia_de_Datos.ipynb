### Pregunta 2
Haz un `merge` de ambos **data frames**, para generar un nuevo **data frame** que incluya los datos de `df_modelos` y `df_marcas`. Tu nuevo **data frame** recibirá el nombre de `df`.

### Pregunta 5
Haz un histograma (`hist`) de la columna `ROM_GB` del **data frame** `df`.

El gráfico debe incluir un título y las columnas tener un color personalizado.

```python
import math
df_ROMGB = df[df['ROM_GB'] == 64]
num_observaciones = df_ROMGB.shape[0]
num_bins = int(math.sqrt(num_observaciones))
plt.figure(figsize=(8,4), dpi=100)
plt.hist(df['ROM_GB'],bins = num_bins,color = '#67c2d3',alpha = 0.7,edgecolor = '#3d3b65', label = 'Datos')
plt.grid(True)
plt.xlabel('Valores de GB en ROM')
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.legend()
plt.show()
```

### Pregunta 6
Haz un diagrama de dispersión (`scatter`) con las siguientes características: el eje X deberá indicar el contenido de la columna `Precio de lista`, mientras que el eje Y expresará el contenido de la columna `Precio de venta`.

El diagrama debe incluir un título y los puntos tener un color personalizado.

```python
x = df['Precio de lista']
y = df['Precio de venta']
plt.figure(figsize=(8,4), dpi=100)
plt.scatter(x,y,s = 100,color = '#67c2d3',alpha = 0.7,edgecolor = '#3d3b65', label = 'Datos')
plt.grid(True)
plt.xlabel('Precio de Lista')
plt.ylabel('Precio de Venta')
plt.title('Diagrama de dispersión')
plt.legend()
plt.show()
```

### Pregunta 8
Utiliza el código para obtener la siguiente información (ubicada en `df`).

- Promedio de los precios de venta de los modelos marca **Realme**
- Máximo de estrellas de los modelos marca **Redmi**
- Mínimo de Estrellas de los modelos marca **Redmi**

```python
columna_filtrada_precio = df[df['Nombre Marca'] == 'Realme']['Precio de venta']
columna_filtrada_estrellas = df[df['Nombre Marca'] == 'Redmi']['Estrellas']
print('El promedio de los precios de venta de los modelos marca Realme es:',columna_filtrada_precio.mean())
print('El máximo de estrellas de los modelos marca Redmi', columna_filtrada_estrellas.max())
print('El mínimo de estrellas de los modelos marca Redmi', columna_filtrada_estrellas.min())
```

### Pregunta 9
Utiliza un **Groupby** para mostrar la cantidad de estrellas promedio que cada marca tiene en `df`.

```python
df.groupby(by='Nombre Marca')['Estrellas'].mean()
```

### Pregunta 10
¿Qué otro dato puedes obtener de `df` y cómo lo extraerías?

Podemos obtener las calificaciones de los modelos y agruparlos por marcas denotando la de mayor calificación en un grafico de pastel o pie.

```python
categorias = ', '.join(df['Nombre Marca'].unique()).split(',')
categorias_ordenadas = sorted(categorias)
df_calificaciones = df.groupby(by='Nombre Marca')['Calificaciones'].mean() * 100
porcentajes = df_calificaciones.values * 100
print(len(categorias_ordenadas))
plt.figure(figsize=(12,8), dpi=100)
_, text, _ = plt.pie(porcentajes,labels=categorias_ordenadas, autopct='%1.1f%%', startangle=90,textprops={'rotation': 'vertical'})
for t in text:
    t.set_fontsize(10)
plt.title('Distribución de Calificaciones')
plt.show()
```
